```python
class OSMParser:
    
    """
    A class to parse OSM XML data to extract nodes, ways, and relations.
    """

    def __init__(self, file_path):
        self.file_path = file_path
        self.text_lines = self._read_file()

    def _read_file(self):
        try:
            with open(self.file_path, 'r', encoding='utf-8') as file:
                lines = file.readlines()
                if not lines:
                    raise ValueError("The file is empty.")
                return lines
        except FileNotFoundError:
            print(f"Error: File not found: {self.file_path}")
            return []
        except ValueError as e:
            print(f"Error: {e}")
            return []

    @staticmethod
    def trim(line):
        return line.strip()

    @staticmethod
    def is_element_start(tag, line):
        return re.match(rf'^.*<{tag}[\s,>].*$', line, flags=re.IGNORECASE) is not None

    @staticmethod
    def is_element_end(tag, line):
        return re.match(rf'^.*</{tag}>.*$', line, flags=re.IGNORECASE) is not None

    @staticmethod
    def extract_attribute(attr, line):
        """Extract a numeric or string attribute value from a line."""
        match = re.search(rf'\s+{attr}="([^"]+)"', line)
        return match.group(1) if match else None

    def extract_node_attributes(self, line):
        """Extract node attributes (id, lon, lat) from a line."""
        try:
            node_id = self.extract_attribute("id", line)
            lon = self.extract_attribute("lon", line)
            lat = self.extract_attribute("lat", line)
            if not all([node_id, lon, lat]):
                raise ValueError("Missing node attributes.")
            return int(node_id), float(lat), float(lon)
        except ValueError as e:
            print(f"Error parsing node attributes: {e}")
            return None, None, None

    def parse_elements(self, tag):
        refs = []
        data = []
        element_open = False
        current_data = ''
        element_id = None

        for line in self.text_lines:
            if self.is_element_start(tag, line):
                element_open = True
                element_id = self.extract_attribute("id", line)
                if not element_id:
                    print(f"Warning: Missing ID for {tag}.")
                    continue
                refs.append(int(element_id))
                current_data = ''
            elif self.is_element_end(tag, line):
                if element_open:
                    data.append(current_data)
                    element_open = False
            elif element_open:
                current_data += self.trim(line)
        return refs, data

    def parse(self):
        # Extract nodes
        node_refs = []
        refs_no_data = []  # Nodes with no tags
        refs_with_data = []  # Nodes with tags
        node_lons = []
        node_lats = []
        refs_data = []  # Metadata for nodes with tags
        node_open = False
        current_node_data = {}

        for line in self.text_lines:
            if self.is_element_start("node", line):
                node_open = True
                node_id, lat, lon = self.extract_node_attributes(line)
                if node_id is not None:
                    node_refs.append(node_id)
                    node_lons.append(lon)
                    node_lats.append(lat)
                    current_node_data = {"id": node_id, "tags": {}}
                # Handle self-closing nodes
                if "/>" in line:
                    node_open = False
                    refs_no_data.append(node_id)
            elif self.is_element_end("node", line):
                node_open = False
                # Classify the node based on whether it has tags
                if current_node_data["tags"]:
                    refs_with_data.append(current_node_data["id"])
                    refs_data.append(current_node_data["tags"])
                else:
                    refs_no_data.append(current_node_data["id"])
            elif node_open and re.search(r'<tag ', line):
                # Extract tags within the node
                key = self.extract_attribute("k", line)
                value = self.extract_attribute("v", line)
                if key and value:
                    current_node_data["tags"][key] = value


        # Extract ways and relations
        way_refs, way_data = self.parse_elements("way")
        rel_refs, rel_data = self.parse_elements("relation")

        return {
            "nodes": {
                "refs": np.array(refs_no_data, dtype=np.int64),
                "refs_with_data": np.array(refs_with_data, dtype=np.int64),
                "lons": np.array(node_lons, dtype=np.float64),
                "lats": np.array(node_lats, dtype=np.float64),
                "data": refs_data,  # Metadata for nodes with tags
            },
            "ways": {
                "refs": np.array(way_refs, dtype=np.int64),
                "data": way_data,
            },
            "relations": {
                "refs": np.array(rel_refs, dtype=np.int64),
                "data": rel_data,
            }
        }



```

```python
import xml.etree.ElementTree as ET
import matplotlib.pyplot as plt
from shapely.geometry import LineString, Point
from collections import defaultdict

# File paths
input_file = "../TpHCM.osm"
filtered_file = "../District_7_filtered.osm"

def filter_district_7_by_relation(input_file, relation_id, output_file):
    """
    Filter District 7 by its relation ID and save the filtered nodes, ways, and relations to a new OSM file.
    """
    # Parse the OSM file
    tree = ET.parse(input_file)
    root = tree.getroot()

    # Locate the target relation
    target_relation = None
    for relation in root.findall("relation"):
        if relation.get("id") == str(relation_id):
            target_relation = relation
            break

    if not target_relation:
        print(f"Relation with ID {relation_id} not found.")
        return

    # Collect references
    way_refs = set()
    node_refs = set()
    relation_refs = set()

    # Collect ways and sub-relations from the target relation
    for member in target_relation.findall("member"):
        if member.get("type") == "way":
            way_refs.add(member.get("ref"))
        elif member.get("type") == "node":
            node_refs.add(member.get("ref"))
        elif member.get("type") == "relation":
            relation_refs.add(member.get("ref"))

    # Create filtered OSM structure
    filtered_root = ET.Element("osm", attrib=root.attrib)

    # Add the target relation
    filtered_root.append(target_relation)

    # Add referenced relations
    for relation in root.findall("relation"):
        if relation.get("id") in relation_refs:
            filtered_root.append(relation)

    # Add ways and collect their node references
    for way in root.findall("way"):
        if way.get("id") in way_refs:
            filtered_root.append(way)
            for nd in way.findall("nd"):
                node_refs.add(nd.get("ref"))

    # Add nodes
    district_7_nodes = {}
    for node in root.findall("node"):
        if node.get("id") in node_refs:
            filtered_root.append(node)
            district_7_nodes[node.get("id")] = (
                float(node.get("lat")),
                float(node.get("lon")),
            )

    # Save the filtered data to a new file
    filtered_tree = ET.ElementTree(filtered_root)
    filtered_tree.write(output_file, encoding="utf-8", xml_declaration=True)

    # Collect boundary lines for plotting
    boundary_lines = []
    for way in root.findall("way"):
        if way.get("id") in way_refs:
            way_points = [
                Point(district_7_nodes[nd.get("ref")][1], district_7_nodes[nd.get("ref")][0])
                for nd in way.findall("nd")
                if nd.get("ref") in district_7_nodes
            ]
            if len(way_points) > 1:
                boundary_lines.append(LineString(way_points))

    return boundary_lines

def plot_boundary(boundary_lines):
    """
    Plot the boundary of District 7 using the extracted boundary lines.
    """
    plt.figure(figsize=(10, 10))
    for line in boundary_lines:
        x, y = line.xy
        plt.plot(x, y, linewidth=1)
    plt.title("Quận 7 Boundary")
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.grid(True)
    plt.show()

# Define relation ID for District 7
district_7_relation_id = 2764875

# Execute filtering
boundary_lines = filter_district_7_by_relation(input_file, district_7_relation_id, filtered_file)

if boundary_lines:
    print(f"Filtered data for Quận 7 saved to: {filtered_file}")
    plot_boundary(boundary_lines)
else:
    print("No boundary lines extracted for plotting.")
```

```python
import re
import numpy as np

def get_osm(file_path):
    # Helper functions to trim whitespaces
    def trim_l(line):
        return re.sub(r'^\s+', '', line)

    def trim_r(line):
        return re.sub(r'\s+$', '', line)

    def trim(line):
        return trim_l(trim_r(line))

    # Check if the line is the start of a node
    def is_node_s(a):
        return re.match(r'^.*<node[\s+,>].*$', a, flags=re.IGNORECASE) is not None

    # Check if the line is the end of a node
    def is_node_e(a):
        return re.match(r'^.*</node>.*$', a, flags=re.IGNORECASE) is not None

    # Check if the line is the start of an element (way or relation)
    def is_elem_s(tag, a, flags=re.IGNORECASE):
        return re.match(r'^.*<' + str(tag) + '[\s+,>].*$', a, flags=flags) is not None

    # Check if the line is the end of an element (way or relation)
    def is_elem_e(tag, a, flags=re.IGNORECASE):
        return re.match(r'^.*</' + str(tag) + '>.*$', a, flags=flags) is not None

    # Extract node ID, latitude, and longitude from a line
    def getnode_id_lon_lat(line):
        xid = re.sub(r'^.*\s+id="([-+]?[0-9]*\.?[0-9]+([eE][-+]?[0-9]+)?)"\s+.*$', r'\1', line)
        xlon = re.sub(r'^.*\s+lon="([-+]?[0-9]*\.?[0-9]+([eE][-+]?[0-9]+)?)"\s+.*$', r'\1', line)
        xlat = re.sub(r'^.*\s+lat="([-+]?[0-9]*\.?[0-9]+([eE][-+]?[0-9]+)?)"\s+.*$', r'\1', line)
        return int(xid), float(xlat), float(xlon)

    # Extract ID from an element (node, way, relation)
    def getelem_id(line):
        xid = re.sub(r'^.*\s+id="([-+]?[0-9]*\.?[0-9]+([eE][-+]?[0-9]+)?)"\s+.*$', r'\1', line)
        return int(xid)

    # Extract tag data from a line
    def gettag_int(tag, line):
        xid = re.sub(r'^.*\s+' + str(tag) + '="([-+]?[0-9]*)"\s+.*$', r'\1', line)
        return int(xid)

    # Open and read the local file
    with open(file_path, 'r', encoding='utf-8') as file:
        Text = file.read().splitlines()  # Read all lines and split by newline
    
    # 1. To extract "node" information (id, lat, lon) and data (id_ref, tags)
    node_ref = []
    node_lon = []
    node_lat = []
    node_ref1 = []
    node_info = []

    nl = 0
    flag_s = flag_e = False
    xid = xlat = xlon = 0
    xdata = ''
    flag_open = False

    for line in Text:
        nl = nl + 1
        flag_s = is_node_s(line)
        flag_e = is_node_e(line)
        if flag_s:
            flag_open = True
            xid, xlat, xlon = getnode_id_lon_lat(line)
            node_ref.append(xid)
            node_lat.append(xlat)
            node_lon.append(xlon)
            xdata = ''
        elif flag_e:
            flag_open = False
            node_ref1.append(xid)
            node_info.append(xdata)
        else:
            if flag_open:
                xdata = xdata + str(trim(line))

    print(f"{1:4d} {'Reading Nodes over No. lines':40} {nl:15d}")
    print(f"{'No. Nodes =':45s} {len(node_ref):15d}")

    # 2. To extract "way" information (id) and way data (node_ref, tags, etc)
    way_ref = []
    way_info = []

    nl = 0
    flag_s = flag_e = False
    wid = 0
    wdata = ''
    flag_open = False

    for line in Text:
        nl = nl + 1
        flag_s = is_elem_s('way', line)
        flag_e = is_elem_e('way', line)
        if flag_s:
            flag_open = True
            wid = getelem_id(line)
            way_ref.append(wid)
            wdata = ''
        elif flag_e:
            flag_open = False
            way_info.append(wdata)
        else:
            if flag_open:
                wdata = wdata + str(trim(line))

    print(f"{2:4d} {'Reading Way over No. lines':40} {nl:15d}")
    print(f"{'No. Way =':45s} {len(way_ref):15d}")

    # 3. To extract "relation" information (id) and data (ref, tags, etc)
    rel_ref = []
    rel_info = []

    nl = 0
    flag_s = flag_e = False
    rid = 0
    rdata = ''
    flag_open = False

    for line in Text:
        nl = nl + 1
        flag_s = is_elem_s('relation', line)
        flag_e = is_elem_e('relation', line)
        if flag_s:
            flag_open = True
            rid = getelem_id(line)
            rel_ref.append(rid)
            rdata = ''
        elif flag_e:
            flag_open = False
            rel_info.append(rdata)
        else:
            if flag_open:
                rdata = rdata + str(trim(line))

    print(f"{3:4d} {'Reading Relation over No. lines':40} {nl:15d}")
    print(f"{'No. Relation =':45s} {len(rel_ref):15d}")

    return np.array(node_ref, dtype='int64'), \
           np.array(node_ref1, dtype='int64'), \
           np.array(way_ref, dtype='int64'), \
           np.array(rel_ref, dtype='int64'), \
           np.array(node_lon, dtype='float64'), \
           np.array(node_lat, dtype='float64'), \
           node_info, \
           way_info, \
           rel_info
```

```python
# get index from ref 
def find_id_from_ref ( aref, iref, getsorted=True ):
    from numpy import argsort, searchsorted
#    
    if ( getsorted ):
        k = searchsorted( aref, iref )
    else:
        ia = argsort( aref )   
        k = searchsorted( aref, iref, sorter=ia )
#        
    return k
```

```python
import folium
import numpy as np
import re  # Import the re library for regex

# Function to extract node references from way data using regex
def extract_node_references_from_way(w_data):
    node_refs = re.findall(r'<nd ref="(\d+)"/>', w_data)
    node_refs = [int(ref) for ref in node_refs]
    return node_refs

def extract_boundary_coords_from_way(t_all_nodes, way_refs):
    boundary_coords = []

    for way_ref in way_refs:
        i_w = find_id_from_ref(ways['refs'], way_ref) 
        w_data = ways['data'][i_w]  
        
        node_refs = extract_node_references_from_way(w_data)

        # Retrieve coordinates for each node reference
        for node_ref in node_refs:
            i_n = t_all_nodes.where('ref', are.equal_to(node_ref))
            if i_n.num_rows > 0:
                lat = i_n.column('lat')[0]  # Extract the latitude
                lon = i_n.column('lon')[0]  # Extract the longitude
                boundary_coords.append([lat, lon])

    return boundary_coords

# Function to plot the boundary on a folium map
def plot_boundary_on_map(center_lat, center_lon, boundary_coords):
    # Create a folium map centered on the centroid (or center point of the district)
    m = folium.Map(location=[center_lat, center_lon], zoom_start=13)

    # Plot the boundary as a polygon (assuming the boundary is closed)
    if boundary_coords:
        folium.Polygon(locations=boundary_coords, color='blue', fill=True, fill_color='blue', fill_opacity=0.3).add_to(m)

    # Display the map
    m.save("district_boundary_map.html")

# Example usage:
id_to_find = 2764875
i = find_id_from_ref(relations['refs'], id_to_find)
way_refs = relations['data'][i] 
way_refs = re.findall(r'<member type="way" ref="(\d+)"', way_refs)

# Convert the ref values to integers
way_refs = [int(ref) for ref in way_refs]

# Extract the boundary coordinates from the way references
boundary_coords = extract_boundary_coords_from_way(t_all_nodes, way_refs)

# Calculate the center of the district (average of the boundary coordinates)
if boundary_coords:
    center_lat = np.mean([lat for lat, lon in boundary_coords])
    center_lon = np.mean([lon for lat, lon in boundary_coords])

    # Plot the boundary on the map
    plot_boundary_on_map(center_lat, center_lon, boundary_coords)

    print(f"Extracted {len(boundary_coords)} coordinates for the boundary.")
else:
    print("No coordinates found for the boundary.")
```
